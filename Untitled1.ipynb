```python
import pandas as pd

# Replace these with your actual file paths or URLs
census_df = pd.read_csv(r"C:\Users\shiva\Downloads\CCD.csv")
schools_df = pd.read_csv(r"C:\Users\shiva\Downloads\ChicagoPublicSchools.csv")
crime_df = pd.read_csv(r"C:\Users\shiva\Downloads\ChicagoCrimeData.csv")
```

```python
# Use SQLAlchemy to connect and push data to SQLite
from sqlalchemy import create_engine
engine = create_engine('sqlite:///FinalDB.db')

# Load the dataframes into SQLite as tables
census_df.to_sql("CCD", con=engine, if_exists="replace", index=False)
schools_df.to_sql("CHICAGOPUBLICSCHOOLS", con=engine, if_exists="replace", index=False)
crime_df.to_sql("CHICAGOCRIMEDATA", con=engine, if_exists="replace", index=False)
```
